```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from utility import get_sample_df
```

```python
df = get_sample_df(path = "/results/autoencoder_fully_connected/realTweets/", file = "autoencoder_fully_connected_Twitter_volume_UPS.csv")
```

```python
df.columns
```

```python
time_ser = df.anomaly_score.values
mu,std = norm.fit(time_ser)
plt.hist(time_ser, bins=10, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
```

```python
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

plt.show()
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


# Generate some data for this demonstration.
data = norm.rvs(10.0, 2.5, size=500)

# Fit a normal distribution to the data:
mu, std = norm.fit(data)

# Plot the histogram.
plt.hist(data, bins=25, density=True, alpha=0.6, color='g')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

plt.show()
```
